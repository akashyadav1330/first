```python
x=np.array([1,2,3,4,5])
y=np.array([2,5,6,9,10,])
mt.title("Line Plot",fontsize=10,color="green",fontweight="bold")
mt.xlabel("x-axis",color="green")
mt.ylabel("y-axis",color="green")
mt.plot(x,marker='*',markersize=10,markerfacecolor='red',markeredgecolor='red')
mt.plot(y,marker='*',markersize=10,markerfacecolor='red',markeredgecolor='red')

mt.show
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,5,6,9,10,])
mt.title("Line Plot",fontsize=10,color="green",fontweight="bold")
mt.xlabel("x-axis",color="green")
mt.ylabel("y-axis",color="green")
mt.grid()
mt.plot(x,marker='*',markersize=10,markerfacecolor='red',markeredgecolor='red')
mt.plot(y,marker='*',markersize=10,markerfacecolor='red',markeredgecolor='red')

mt.show
```

scatter

```python
x=np.array([1,2,3,4,5])
y=np.array([2,5,6,9,10,])
mt.title("Line Plot",fontsize=10,color="green",fontweight="bold")
mt.xlabel("x-axis",color="green")
mt.ylabel("y-axis",color="green")
mt.grid()
mt.scatter(x,y)

mt.show
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,5,6,9,10,])
c=np.array(["red","green","blue","yellow","pink"])
size=np.array([500,300,800,350,270])
mt.title("Line Plot",fontsize=10,color="green",fontweight="bold")
mt.xlabel("x-axis",color="green")
mt.ylabel("y-axis",color="green")
mt.grid()
mt.scatter(x,y,c=c,s=size)

mt.show
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
mt.scatter(x, y,alpha=0.4, s = sizes, c = c ,cmap='viridis')# map, viridis, nipy_spectral, coolwarm
mt.colorbar()
mt.show()
```
